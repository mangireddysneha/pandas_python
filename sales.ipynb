```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales.csv', encoding='latin-1')

print("🔍 First 5 rows:")
print(df.head())

print("\n📊 Dataset Info:")
print(df.info())

sales_by_product = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()
print("\n📈 Total Sales by Product:")
print(sales_by_product)

sales_by_region = df.groupby('COUNTRY')['SALES'].sum().reset_index()
print("\n📌 Total Sales by Region:")
print(sales_by_region)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='PRODUCTLINE', y='SALES', data=sales_by_product, palette='Blues_d')
plt.title("Total Sales by Product")
plt.ylabel("Sales Amount")
plt.xlabel("Product Line")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='COUNTRY', y='SALES', data=sales_by_region, palette='Greens_d')
plt.title("Total Sales by Region")
plt.ylabel("Sales Amount")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```
